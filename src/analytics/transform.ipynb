### Fee data in transactions on Ethereum Blocks 

```python
transaction_data.info()
```

```python
feeData = transaction_data[['blockNumber', 'transaction', 'gasUsed', 'gasPrice']].copy()
feeData.sample(5)
```

```python
# gasPrice [Ether] convert to gasPrice [USD]
gasPriceUSD = feeData['gasPrice'] / 1000000000000000000
# 1 USD = 33.48 THB
gasPriceTHB = gasPriceUSD * float(33.48)
# Value of Network Fee structure > totalFee = networkFee + itx 
# Value of Network Fee structure unit is Ether
networkFee = feeData['gasUsed'] * feeData['gasPrice'] 
itx = feeData['gasUsed'] * 0.00000002  # Gas Price: 0.00000002 Ether (20 Gwei)
totalFee = networkFee + itx 

# Change units from 'Ether' to 'USD' and 'THB' 
# unit is USD
# 1 ETH = 3863.66 USD
totalFee_priceUSD = totalFee / 1000000000000000000

# unit is THB
# 1 ETH = 129,355.34 THB
# 1 USD = 33.48 THB
totalFee_priceTHB = totalFee_priceUSD * float(33.48)
```

```python
feeData['networkFee'] = networkFee
feeData['itx_commission'] = itx
feeData['totalFeePrice'] = totalFee
feeData['totalFeePriceUSD'] = totalFee_priceUSD
feeData['totalFeePriceTHB'] = totalFee_priceTHB
```
